```python
# Count frequent words in skill descriptions
from collections import Counter

# Combine all skill descriptions into one string
all_skills = " ".join(df['Skill description'].dropna().astype(str))

# Break into words
words = all_skills.lower().split()

# Count words
word_counts = Counter(words)

# Show most common ones
common_words = word_counts.most_common(20)
print("Most common words in skill descriptions:")
for word, count in common_words:
    print(f"{word}: {count}")
```

```python
import matplotlib.pyplot as plt

# Define boring common words to ignore
stop_words = {'the', 'it', 'will', 'can', 'be', 'to', 'and', 'in', 'you', 'if', 'is', 'of', 'on', 'its', 'while', 'around', 'player\'s', 'back.', 'this', 'a'}

# Filter the useful words
filtered_words = [(word, count) for word, count in word_counts.items() if word not in stop_words and len(word) > 3]

# Sort and keep top 15
filtered_words = sorted(filtered_words, key=lambda x: x[1], reverse=True)[:15]

# Plot
words, counts = zip(*filtered_words)

plt.figure(figsize=(10, 5))
plt.bar(words, counts, color='green')
plt.title("🔧 Most Common Work-Related Keywords")
plt.xlabel("Keyword")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Show the least common useful words (filtered same as before)
least_common = sorted(filtered_words, key=lambda x: x[1])[:15]

# Plot
words, counts = zip(*least_common)

plt.figure(figsize=(10, 5))
plt.bar(words, counts, color='red')
plt.title("🧪 Least Common Work-Related Keywords")
plt.xlabel("Keyword")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Define attributes to assess versatility
versatile_attrs = ['HP', 'melee attack', 'Remote attack', 'defense', 'support']

# Drop rows with missing values in these columns
df_versatile = df.dropna(subset=versatile_attrs).copy()

# Normalize (scale between 0 and 1)
scaler = MinMaxScaler()
df_versatile_scaled = pd.DataFrame(scaler.fit_transform(df_versatile[versatile_attrs]), columns=versatile_attrs)

# Add total versatility score
df_versatile['Versatility Score'] = df_versatile_scaled.sum(axis=1)

# Top 10 most versatile Pals
top_versatile = df_versatile.sort_values(by='Versatility Score', ascending=False).head(10)

# Show results
top_versatile[['Name', 'Element 1', 'HP', 'melee attack', 'Remote attack', 'defense', 'support', 'Versatility Score']]
```

```python
# Plot top 10 versatile Pals
plt.figure(figsize=(10, 5))
sns.barplot(data=top_versatile, x='Name', y='Versatility Score', palette='viridis')
plt.title("💪 Top 10 Most Versatile Pals")
plt.xlabel("Pal Name")
plt.ylabel("Versatility Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Define columns that matter for exploration
exploration_attrs = ['walking speed', 'running speed', 'Riding sprint speed', 'endurance', 'support']

# Drop rows with missing values
df_explore = df.dropna(subset=exploration_attrs).copy()

# Normalize
scaler = MinMaxScaler()
df_explore_scaled = pd.DataFrame(scaler.fit_transform(df_explore[exploration_attrs]), columns=exploration_attrs)

# Add exploration score
df_explore['Exploration Score'] = df_explore_scaled.sum(axis=1)

# Top 10 explorers
top_explorers = df_explore.sort_values(by='Exploration Score', ascending=False).head(10)

# Show result
top_explorers[['Name', 'Element 1', 'walking speed', 'running speed', 'Riding sprint speed', 'endurance', 'support', 'Exploration Score']]
```

```python
# Bar chart for top explorers
plt.figure(figsize=(10, 5))
sns.barplot(data=top_explorers, x='Name', y='Exploration Score', palette='coolwarm')
plt.title("🧭 Top 10 Pals for Exploration")
plt.xlabel("Pal Name")
plt.ylabel("Exploration Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
