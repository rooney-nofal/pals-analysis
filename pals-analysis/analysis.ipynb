```python
df = pd.read_csv("Palworld_Data--Palu combat attribute table.csv", skiprows=1)
df.head()
```

```python
# Re-check missing values with proper columns
df.isnull().sum()
```

```python
# Drop unnecessary or less useful columns with heavy missing data


# Fill missing 'Element 2' with 'None' (or could use Element 1 if you prefer)
df['Element 2'] = df['Element 2'].fillna('None')

# Fill 'variant' and 'nocturnal' with appropriate defaults
df['variant'] = df['variant'].fillna('Base')
df['nocturnal'] = df['nocturnal'].fillna(False)

# Fill missing sprint speed with 0
df['Riding sprint speed'] = df['Riding sprint speed'].fillna(0)
```

```python
print(df.columns.tolist())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='rarity', palette='viridis')

# Add labels and title
plt.title('Distribution of Pal Rarity')
plt.xlabel('Rarity')
plt.ylabel('Number of Pals')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Volume size', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Pal Volume Size')
plt.xlabel('Volume Size')
plt.ylabel('Number of Pals')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='HP', bins=20, kde=True, color='salmon')
plt.title('Distribution of Pals\' HP')
plt.xlabel('HP (Health Points)')
plt.ylabel('Number of Pals')
plt.tight_layout()
plt.show()
```

```python
df['Combat Power'] = df['melee attack'] + df['Remote attack'] + df['defense']
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Combat Power', bins=20, kde=True, color='purple')
plt.title('Distribution of Combat Power')
plt.xlabel('Combat Power (melee + remote + defense)')
plt.ylabel('Number of Pals')
plt.tight_layout()
plt.show()
```

```python
# Sort the dataframe by Combat Power in descending order
top_10 = df.sort_values(by='Combat Power', ascending=False).head(10)

# Display name, tribe, rarity, and combat stats
top_10[['Name', 'Tribe', 'rarity', 'melee attack', 'Remote attack', 'defense', 'Combat Power']]
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='GenusCategory', order=df['GenusCategory'].value_counts().index, palette='coolwarm')
plt.title('Distribution of Pal Categories')
plt.xlabel('Category (Genus)')
plt.ylabel('Number of Pals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='endurance', bins=15, kde=True, color='green')
plt.title("Distribution of Pal Endurance (Food Consumption Proxy)")
plt.xlabel("Endurance")
plt.ylabel("Number of Pals")
plt.tight_layout()
plt.show()
```

```python
# Look for any skill descriptions that mention useful resources
resource_keywords = ['milk', 'egg', 'wool', 'cotton', 'fiber', 'produce']

# Filter rows where any keyword appears in Skill description (case-insensitive)
resource_pals = df[df['Skill description'].str.lower().str.contains('|'.join(resource_keywords), na=False)]

# Show relevant columns
resource_pals[['Name', 'Tribe', 'rarity', 'Skill description']]
```

```python
# Select relevant combat/stat attributes
combat_cols = ['HP', 'melee attack', 'Remote attack', 'defense', 'support', 'Speed of work', 'endurance', 'Combat Power']

# Compute correlation matrix
correlation_matrix = df[combat_cols].corr()

# Plot heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Combat Attributes')
plt.tight_layout()
plt.show()
```
